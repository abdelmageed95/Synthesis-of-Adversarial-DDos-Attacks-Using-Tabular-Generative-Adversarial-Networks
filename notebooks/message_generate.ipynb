```python
import pandas as pd 
df = pd.read_csv("fake_data.csv")
df
```

```python
names = df.columns
for name in names:
    df[name] = df[name].astype("str")
```

```python
data = []
for r in range(300):
    data_row = []
    for i in range(len(names)):
        data_row.append(df.iloc[r,:][i])
    data.append(data_row)

tests = []
for ele in data:
        txt = ','.join(ele)
        tests.append(txt)

dic = {"row_data" :tests}
fake_data = pd.DataFrame(dic)
fake_data.head()
```
